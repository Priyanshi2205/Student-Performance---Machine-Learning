```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#here survived is the independent variable thus it is taken in another variable 
X = df[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid',
       'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2',
       'G3']]
y = df[['pass']]
y
```

```python
#to split the data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25)
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

**Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train_norm = independent_scaler.fit_transform(X_train)
X_test_norm = independent_scaler.transform(X_test)
```

```python
X_train_norm
```

```python
X_test_norm
```

**PCA**

```python
#PCA - Principal component Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components =2)              # Convert All the columns into two components/columns
X_train_pca = pca.fit_transform(X_train_norm)
X_train_pca =pca.transform(X_train_norm)
X_test_pca = pca.fit_transform(X_test_norm)
X_test_pca =pca.transform(X_test_norm)
pca.explained_variance_ratio_
```

```python
import math
math.sqrt(len(y_test))
```

**KNN Algorithm**

```python
#fromâ€¯sklearn.neighborsâ€¯importâ€¯KNeighborsClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report ,confusion_matrix ,ConfusionMatrixDisplay
```

```python
# Define the model: Init K-NN
knn = KNeighborsClassifier(n_neighbors = 11, p=2,metric='euclidean')
```

```python
# Fit Model
knn.fit(X_train_pca , y_train)
```

```python
# Predict the test set results
y_pred =knn.predict(X_test_pca)
y_pred
```

```python
# Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print (cm)
target_names = ['passed','not passed']
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_score
print(precision_score(y_test, y_pred))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
